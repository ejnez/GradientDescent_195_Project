```python
#imports
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# 1. Load the preprocessed dataset
df = pd.read_csv('Model_merged_ABC_2010_2019.csv')
y = df['wheat_yield_bu_acre']
```

```python
# 2. Define the Feature Sets
climate_features = ['mean_summer_temp_C', 'total_growing_precip_mm', 
                    'frost_free_days', 'spring_temp_anomaly_C']
bryophyte_features = ['richness_per_site', 'ids_per_site']

X_baseline = df[climate_features]
X_augmented = df[climate_features + bryophyte_features]
```

```python
# 3. Build and Fit Model 1 (Baseline)
model_baseline = LinearRegression()
model_baseline.fit(X_baseline, y)
predictions_baseline = model_baseline.predict(X_baseline)
# Calculate RMSE for Model 1
# RMSE represents the average distance between predicted yields and actual yields.
rmse_baseline = np.sqrt(mean_squared_error(y, predictions_baseline))
print(f"Baseline Climate Model RMSE: {rmse_baseline:.2f} bushels/acre")
```

```python
# 4. Build and Fit Model 2 (Augmented)
model_augmented = LinearRegression()
model_augmented.fit(X_augmented, y)
predictions_augmented = model_augmented.predict(X_augmented)

rmse_augmented = np.sqrt(mean_squared_error(y, predictions_augmented))
print(f"Augmented Eco-Climate Model RMSE: {rmse_augmented:.2f} bushels/acre")
```
