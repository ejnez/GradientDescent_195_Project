```python
#imports
import pandas as pd
import numpy as np
import os
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
def adjusted_r2(X, y, predictions):
    r2 = r2_score(y, predictions)
    n = len(y)
    p = X.shape[1]
    return 1 - (1 - r2) * (n - 1) / (n - p - 1)
```

```python
DATA_DIR = "preprocessing_results"
DATA_FULL_FILE = "Model_merged_ABC_2008_2010_2019.csv"
```

```python
# 1. Load the preprocessed dataset
df = pd.read_csv(os.path.join(DATA_DIR, DATA_FULL_FILE))
y = df['wheat_yield_bu_acre']
```

```python
# 2. Define the Feature Sets
climate_features = ['mean_summer_temp_C', 'total_growing_precip_mm', 
                     'spring_temp_anomaly_C']
bryophyte_features = ['richness_per_100min']

X_A = df[climate_features] # baseline
X_B = df[climate_features + bryophyte_features] # augmented
X_C = df[bryophyte_features]
```

```python
# 3. Build and Fit Model A (Baseline)
model_A = LinearRegression()
model_A.fit(X_A, y)
predictions_A = model_A.predict(X_A)
# Calculate RMSE for Model 1
# RMSE represents the average distance between predicted yields and actual yields.
rmse_A = np.sqrt(mean_squared_error(y, predictions_A))
print(f"Baseline Climate Model RMSE: {rmse_A:.2f} bushels/acre")
```

```python
# 4. Build and Fit Model C (Augmented)
model_B = LinearRegression()
model_B.fit(X_B, y)
predictions_B = model_B.predict(X_B)

rmse_B = np.sqrt(mean_squared_error(y, predictions_B))
print(f"Augmented Eco-Climate Model RMSE: {rmse_B:.2f} bushels/acre")
```

```python
# 4. Build and Fit Model C (Augmented)
model_C = LinearRegression()
model_C.fit(X_C, y)
predictions_C = model_C.predict(X_C)

rmse_C = np.sqrt(mean_squared_error(y, predictions_C))
print(f"Augmented Eco-Climate Model RMSE: {rmse_C:.2f} bushels/acre")
```

```python
print(model_A.coef_)
print(model_B.coef_)
print(model_C.coef_)

print(adjusted_r2(X_A, y, predictions_A))  # 3 predictors
print(adjusted_r2(X_B, y, predictions_B))  # 4 predictors
print(adjusted_r2(X_C, y, predictions_C))  # 1 predictor
```
